```python
print("Category Data Encoding:")
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
df.iloc[:,3]=m.fit_transform(df.iloc[:,3].values.reshape(-1,1))
df.iloc[:,4]=m.fit_transform(df.iloc[:,4].values.reshape(-1,1))
df.iloc[:,5]=m.fit_transform(df.iloc[:,5].values.reshape(-1,1))
df.iloc[:,6]=m.fit_transform(df.iloc[:,6].values.reshape(-1,1))
df.iloc[:,7]=m.fit_transform(df.iloc[:,7].values.reshape(-1,1))

df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.iloc[:,-1]=le.fit_transform(df.iloc[:,-1])
df
```

```python
print("Handling Missing Values:")
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
df.iloc[:,3]=imp_mean.fit_transform(df.iloc[:,3].values.reshape(-1,1))
df.iloc[:,4]=imp_mean.fit_transform(df.iloc[:,4].values.reshape(-1,1))
df.iloc[:,5]=imp_mean.fit_transform(df.iloc[:,5].values.reshape(-1,1))
df
```
